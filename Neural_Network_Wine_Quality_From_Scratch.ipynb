# Neural Network for Wine Quality From Scratch

Multi-layer neural network implemented from scratch using only NumPy for wine quality prediction.

## Features

- Feedforward network
- Backpropagation
- Manual gradient descent
- ReLU activation
- NumPy only

```python
import pandas as pdimport numpy as npfrom sklearn.model_selection import train_test_splitfrom sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("wine.csv")
```

```python
data.head()
```

```python
X = data.drop('quality', axis=1)
```

```python
y = data["quality"].values.reshape(-1,1)
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state= 42, test_size = 0.2)
```
